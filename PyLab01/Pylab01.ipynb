### How about hyperbolic tangent?

$f(z) = \tanh(z)$

$\tanh(z) = \frac{\sinh(z)}{\cosh(z)} = \frac{e^z - e^{-z}}{e^z+e^{-z}}$

```
z = np.arange(0, 5+0.1, 0.1)

z
```

```
g = np.tanh(z)

g
```

$\omega^2 = gk\tanh(kh)$

```
plt.plot(z, g)
plt.xlabel('kh', fontsize=14)
plt.ylabel('$\omega^2$', fontsize=14)
plt.grid(True)
plt.title('Dispersion Relation for Surface Gravity Waves')
```

## Subplots

Lets put the parabola and hyperbolic tangent functions side-by-side using subplots

```
plt.subplot(2, 2, 1)
plt.plot(x, f)

plt.subplot(2, 2, 2)
plt.plot(z, g)

# plt.subplot(2, 2, 3)

# plt.subplot(2, 2, 4)
```

```
plt.subplot(1, 2, 1)
plt.plot(x, f)

plt.subplot(1, 2, 2)
plt.plot(z, g)
```

Lets use subplots()

```
fig, ax = plt.subplots(1, 2)

ax[0].plot(x, f)
ax[0].set_xlabel('x', fontsize=14)
ax[0].set_ylabel('f(x)', fontsize=14)
ax[0].set_title('Parabola', fontsize=14)

ax[1].plot(z, g)
ax[1].set_xlabel('z', fontsize=14)
ax[1].set_ylabel('g(z)', fontsize=14)
ax[1].set_title('Hyperbolic Tangent', fontsize=14)

```
